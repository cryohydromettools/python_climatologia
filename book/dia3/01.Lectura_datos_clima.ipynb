## Importar librerias

```python
import pandas as pd
```

## Leer archivo

```python
data = pd.read_csv('../../data/qc00000543.txt')
data
```

Podemo usar **`delimiter`** para separar las columnas.

```python
data = pd.read_csv('../../data/qc00000543.txt', delimiter=' ')
data
```

El archivo no tiene encabezado. Entonces podemos usar **`delimiter`** para indicar que el archivo no tiene encabezado.

```python
data = pd.read_csv('../../data/qc00000543.txt', delimiter=' ', header=None)
data
```

Los datos tienen valores perdidos y estos estan declarados como -99.9. Podemos usar 

```python
data = pd.read_csv('../../data/qc00000543.txt', delimiter=' ', header=None, na_values=-99.9)
data
```

Ahora vamos a agragar en nombre de las variables para cada columna. Para eso es importante saber que representa cada columna.

```python
data.columns = ['Year', 'Month', 'Day', 'Prec', 'Tmax', 'Tmin']
data
```

Ahora vamos a gener una variable con formato datetime. Para eso usaremos  **`to_dateime`**.

```python
# Esto es un ejemplo de como generar una variable con formato tiempo.
pd.to_datetime(data[['Year', 'Month', 'Day']])
```

Vamos a ingresar esta variable en nuestro dataframe como el index. Esto hara que python pueda trabajar como una serie temporal.

```python
data.index = pd.to_datetime(data[['Year', 'Month', 'Day']])
data
```

Podemos colocar un nombre a nuestro Index.

```python
data.index.name = 'Datetime'
data
```

De esta forma ya estamos lito para usar nuestro dataframe como una serie de tiempo. Ahora podemos aprovechar una gran variedad de funcione disponibles en **`pandas`** para usar con nuetro dataframe. Por exmplo, podemos usar la función **`plot`** para hacer un grafico rapido de nuestros datos.

```python
data['Tmax'].plot()
```

Podemo explorar nuestros datos en la figura que generamos. 

Podemos seleccionar más de dos variables para hacer una figura.

```python
data[['Tmax', 'Tmin']].plot()
```
